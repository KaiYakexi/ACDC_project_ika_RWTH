# Abstract

*At present, the Kalman filter has made a significant contribution to the research of the estimation of the variables of interest when they can't be measured directly. It not only simplifies the experimental device to study the motion mode to a large extent but also achieves better research results when combined with an optimal estimation algorithm. Before that, Object tracking often relies on multi-instance Kalman filters. While good tracking results can be achieved with a regular Kalman filter, it is limited to linear system models and assumptions of Gaussian noise, s.t. it often fails to predict more complex systems accurately.
In real-world traffic, road users rarely behave linearly. Additionally, the output of some sensors, such as radars, can also be nonlinear. The assumption of Gaussian noise also breaks in the presence of arbitrarily distributed noise. Therefore, more ambitious systems or sensor models require more advanced filtering methods. Therefore, the purpose of this study is first to do some Kalman filter limitation analysis and then develop a better model to use a more robust filter technique for object fusion and tracking.*

*In this report, the Extended Kalman filter (EKF) is selected. The Extended Kalman filter solves the problem that the standard Kalman filter is easy to lose the ability to estimate a more complex level system because of the limitation under the linear model assumption and gaussian noise. This report shows that the standard Kalman filter has low capabilities to estimate under more complex noise. Additionally, the algorithm of the Extended Kalman filter will be provided, and try to implement it under more complex nonlinear models such as Constant Turn Rate and Velocity (CTRV) and Constant Turn Rate and Acceleration (CTRA) with more complex noise.*

# Introduction

*Object fusion and tracking are pinpointed on the left side of the A-model [[1]](https://www.researchgate.net/publication/326976679_Automated_driving_-_Potentials_challenges_and_solutions), where is part of and combines environment modeling and environment perdition. 
In the studies of Effert, Rauskolb et al. the object fusion and tracking can be divided into three different subsections ([[2]](https://link.springer.com/chapter/10.1007/978-3-642-03991-1_11), [[3]](https://leopard.tu-braunschweig.de/receive/dbbs_mods_00029859)). The first step is object prediction, the objects detected by the sensor need to be aligned temporally with global objects so that a motion model can be used to predict the state of the vehicle. Most of relevant studies use Kalman filter [[4]](https://asmedigitalcollection.asme.org/fluidsengineering/article-abstract/82/1/35/397706/A-New-Approach-to-Linear-Filtering-and-Prediction) to perform prediction. The second step is called object association, which means efficiently associating sensor objects to global objects and finding candidates for new global objects subsequently. In the end, object association can point out the undetected global objects as well. Even more to the point are two methods used for measuring the sensor-level object and global object. Intersection over Union and Mahalanobis distance is the association measure for bringing thresholds to accomplish the task. Mahalanobis is a stronger choice than the Euclidean distance as a distance measure because it accounts for the uncertainty of the objects and other variables than position can also be used for distance calculation. The last step is called object fusion, which basically fuses all object states and minimizes uncertainties between states.*

*Object tracking often relies on multi-instance kalman filters. While good tracking results can be achieved with a regular Kalman filter, it is limited to linear system models and assumptions of Gaussian noise, s.t. it often fails to accurately predict more complex systems.*

## *Filtering Techniques State-of-Art in Object Tracking*

*Different filter techniques have been carried out during the past decades. Various computer vision methods are used in single Object tracking and classified by Zahra Soleimanitaleb et al. in 2019. Among those, estimation methods are first shortly categorized, focusing only on current popular methods such as kalmen filers and particle filers [[20]](https://www.semanticscholar.org/paper/Object-Tracking-Methods%3AA-Review-Soleimanitaleb-Keyvanrad/c403478b58d9bfa608101b715194b9c4cfbd6c07). Nevertheless, there have been multiple attempts for both objects detection and tracking in previous studies.*

*Exceptionally, in the application of security, automated surveillance systems are very crucial for detecting and tracking moving objects. Different filter technologies were used for detecting and tracking moving objects, consequently, they got extended to adapted different applications.*

*For illustration of detection used filters, Sam Schauland et al. proposed a digital filter to point out the separation of multi-dimensional signals, which presented good performance for motion-based object during detection phase. [[21]](https://ieeexplore.ieee.org/abstract/document/4594786?casa_token=y9CHL8cPKiMAAAAA:snRHaqJFZfR_hIc1rUJvce4NSc6iYexK7j2mnEyn4Nk9uvKy8MwlXgamzVV1szn2YOIPvCo). ## *Advanced Filters for Non-linear Systems and Noise Robustness*

*Kalman filter is a broadly used optimal status estimation method, which can conclude acceptable tracking results under the assumption of linear systems and Gaussian noise. Nonetheless, in most practical applications, systems are non-linear, and noise is non-gaussian distributed. Under the circumstance of non-linear models, the standard Kalman filter suffers performance degradation. Appropriately, many other prevailing methods have been proposed in recent years to address the state estimation problems under non-linear systems and non-gaussian noise. Li et al. analyzed and compared the advantages and disadvantages of Kalman filter (KF), Extended Kalman filter (EKF), and likewise Unscented Kalman Filter (UKF) in 2015 [[5]](https://ieeexplore.ieee.org/document/7528889). Among those, the particle filter also become a popular choice [[6]](https://citeseerx.ist.psu.edu/viewdoc/download?doi=10.1.1.6.168&rep=rep1&type=pdf). In 2022, Kiani and Ahmadvand recapitulated all possible advanced filters used for non-linear systems in recent years. Different combinations of various kalman filters made significant improvements [[7]](https://ui.adsabs.harvard.edu/abs/2022ITAES..58.3261K/abstract).*

### *Extended Kalman Filter (EKF)*
*EKF makes a local linearization of system equations around the estimated status. A non-linear system problem is transformed into a linear system problem, which can be solved like Kalman Filter [[5]](https://ieeexplore.ieee.org/document/7528889). Although EKF can handle non-linear problems, EKF is still under the assumption of gaussian noise.*

### *Unscented Kalman Filter (UKF)*
*Unscented Kalman Filter is an a distinct prevalent filtering method, which can be applied to non-linear system [[5]](https://ieeexplore.ieee.org/document/7528889). UKF was proposed firstly by Julier and Uhlmann [[8]](https://www.cs.unc.edu/~welch/kalman/media/pdf/Julier1997_SPIE_KF.pdf), and later on be developed to a larger class of methods based on UKF, such as STUKF [[9]](https://dl.acm.org/doi/10.1109/ICASSP.2016.7472470), OUKF, RAUKF ([[10]](https://pubmed.ncbi.nlm.nih.gov/29518960/) , [[11]](https://www.mdpi.com/1424-8220/19/6/1371)). UKF is also under the assumption of gaussian noise, granting all this some developed methods try to describe the noise differently. In addition, Robust Adaptive Unscented Kalman Filter (RAUKF) is a new method was studied for gaussian noise with uncertain noise covariance. Moreover, Huang et al. suggested  STUKF for the non-gaussian heavy-tailed noise in 2016, in which noise was described by using Student distribution.*

### *Particle Filter (PF)*
*The application field of many different Kalman filter variants is limited to gaussian noise. Compared to Kalman filter, the particle filtering method can characterize the nature of non-gaussian distributed noise well. A Particle filter is a flexible and efficient method for non-linear processes and non-gaussian noise. Nevertheless, a Particle filter also has some destructive disadvantages. Thus, if there are not enough particles in the estimation filter, the quality of the status estimation decreases sharply. Additionally, there is possible high demand for computational resources and computational time ([[3]](https://leopard.tu-braunschweig.de/receive/dbbs_mods_00029859), [[12]](https://ieeexplore.ieee.org/document/978374)).*

### *The Smooth Variable Structure Filter (SVSF)*
*SVSF was proposed as an extension of VSF [[13]](https://asmedigitalcollection.asme.org/dynamicsystems/article-abstract/125/3/287/459217/The-Variable-Structure-Filter?redirectedFrom=fulltext) for non-linear smooth dynamic systems with finite uncertainties and gaussian noises [[14]](https://ieeexplore.ieee.org/document/4266879). Later, SVSF was studied with various Kalman filter to form a advanced approach called CV-SVSF, which outperforms any other different combination of extended methods ([[15]](https://www.researchgate.net/publication/262327278_Combined_cubature_Kalman_and_smooth_variable_structure_filtering_A_robust_nonlinear_estimation_strategy), [[16]](https://ieeexplore.ieee.org/document/6850138)).*

### *Sliding innovation Filter (SIF)*
*The innovation filter (SIF) has recently been recommended for gaussian uncertain linear systems. Today, (SIF) has been extended to non-linear systems via calculating Jacobians of the dynamic systems. Meanwhile, studies showed the SIF outperforms the SVSF ([[17]](https://ieeexplore.ieee.org/stamp/stamp.jsp?arnumber=9096294), [[18]](https://iemtronics.org/wp-content/uploads/2020/09/IEMTRONICS-2020_Schedule-4.pdf) ). In 2022, a new robust estimation method is established based on the sliding innovation filter (SIF), which can remove the restriction of requiring the system smoothness and Jacobian calculation in the SIF [[7]](https://ui.adsabs.harvard.edu/abs/2022ITAES..58.3261K/abstract).*

# Methodology

*Before we use the Extended Kalman Filter, we want to play with the noise and use Kalman Filter to see the results of object fusion. Thus, our methodology will including the analysis of Kalman Filter analysis, and afterwards, we will introduce some handwritten model assumptions.We will mostly focus on the CTRV model.*

1. *Clearify the state vector*

    $\boldsymbol{O}_i=\left(\begin{array}{ll}\widehat{\boldsymbol{x}}\quad \boldsymbol{P}\end{array}\right)^T$ *(Object)*
    
    $\widehat{\boldsymbol{x}}=\left(x\quad y\quad z\quad v_x\quad v_y\quad a_x\quad a_y\quad l\quad w\quad h\quad\right)^T$ *(Estimated object state vector)*
    
    $\boldsymbol{P}=\operatorname{cov}\left(\widehat{\boldsymbol{x}}_{e r r}, \widehat{\boldsymbol{x}}_{e r r}\right)$ *(Uncertainty of object state, described by its error covariance matrix)*
    
    $x\quad y\quad z$ *(Bounding box center position)*
    
    $v_x\quad v_y\quad a_x\quad a_y$ *(Velocities and accelerations)*
    
    $l\quad w\quad h$ *(Bounding box length, width, height)*
    
    $\widehat{\boldsymbol{x}}_S$ *(Measured state vector: "S" = sensor)*
    
    $\widehat{\boldsymbol{x}}_G$ *(Global/Fused state vector: "G" = global)*


2. Kalman FIlter

- *Prediction Step*

    $\widehat{\boldsymbol{x}}_G:=\boldsymbol{F} \widehat{\boldsymbol{x}}_G $
    
    $\boldsymbol{P}_G:=\boldsymbol{F} \boldsymbol{P}_G \boldsymbol{F}^T+\boldsymbol{Q}$

   *Corresponding C++ code:*

- *Update Step* 

    $\boldsymbol{K}:=\boldsymbol{P}_G\boldsymbol{C}^T(\boldsymbol{\boldsymbol{C}\boldsymbol{P}_S\boldsymbol{C}^T+\boldsymbol{C}\boldsymbol{P}_G\boldsymbol{C}^T})^{-1}$ 
    
    $\widehat{\boldsymbol{x}}_G:= \widehat{\boldsymbol{x}}_G+ \boldsymbol{K}(\boldsymbol{C} \widehat{\boldsymbol{x}}_S-\boldsymbol{C} \widehat{\boldsymbol{x}}_G)$  
    
    $\boldsymbol{P}_G:=\boldsymbol{P}_G-\boldsymbol{K}\boldsymbol{C}\boldsymbol{P}_G $

3. *Extended Kalman Filter*

*Extended Kalman Filter (EKF) solves the nonlinear problem through local linearity. The nonlinear prediction and observation equations are derived and linearized by replacing them with tangent lines. The first-order Taylor expansion is performed at the mean value.*

*Prediction:*
    
$u_{G}=g\left(\widehat{\boldsymbol{x}}_G, u_{G}\right)$

$\boldsymbol{P}_G:=\boldsymbol{G}_J\boldsymbol{P}_G\boldsymbol{G}_J^T+\boldsymbol{Q}$
    
*Update:*

$\boldsymbol{K}:=\boldsymbol{P}_G\boldsymbol{H}^T(\boldsymbol{\boldsymbol{C}\boldsymbol{P}_S\boldsymbol{C}^T+\boldsymbol{H}\boldsymbol{P}_G\boldsymbol{H}^T})^{-1}$

$\widehat{\boldsymbol{x}}_G:= \widehat{\boldsymbol{x}}_G+ \boldsymbol{K}(\boldsymbol{C} \widehat{\boldsymbol{x}}_S-\boldsymbol{h}(\widehat{\boldsymbol{x}}_G))$

$\boldsymbol{P}_G:=\boldsymbol{P}_G-\boldsymbol{K}\boldsymbol{H}\boldsymbol{P}_G$

*Let Kalman first make a comparison with the extended Kalman filter. It can be seen from the comparison that the extended Kalman is a simple nonlinear approximate filtering algorithm, which refers to the case that the motion or observation equation is not linear. The prediction model and measurement model of the extended Kalman are already nonlinear in the prediction model part. In order to simplify the calculation, EKF linearizes the equations of motion and observation through first-order Taylor decomposition. KF and EKF have the same algorithm structure. They both describe the posterior probability density in the form of Gauss and are obtained by calculating the Bayesian recurrence formula. The most significant difference is that when calculating variance, the state transition matrix of EKF (the state information and observation matrix at the previous time are Jacobian matrices of state information (matrices composed of partial derivatives). In the prediction formula part, $G_J$ of the extended Kalman filter is the Jacobian matrix of g, and in the update formula part, H of the extended Kalman filter is the Jacobian matrix of h.*

*Constant Velocity (CV) and Constant Acceleration (CA) are two commonly used linear models. These linear motion models assume that the target is moving in a straight line without considering the turning of the object. 
Constant Turn Rate and Velocity (CTRV) and Constant Turn Rate and Acceleration (CTRA) are two nonlinear models.
Currently, CTRVs are mostly used for airborne tracking systems (aircraft). Most of these quadratic motion models assume velocity has nothing to do with yaw rate. Therefore, in this kind of motion model, due to the disturbance (instability) of the yaw angular velocity measurement, even if the vehicle does not move, the angular velocity under our motion model will change slightly. The correlation between velocity and yaw rate can be achieved by setting the steering angle to solve this problem. The constant steering angle and velocity (CSAV) model is derived from the constant steering angle and velocity model. In addition, the velocity can be assumed to be linear, and the constant curvature and acceleration (CCA) model is derived [[39]](https://www.mdpi.com/1424-8220/22/15/5828).*

# Evaluation and Discussion

*The extended Kalman filter solves the nonlinear estimation problems by linearising the state during the prediction step and measurement equations during the update step. Technically, the Extended Kalman filter just applies the standard Kalman filter formulas to the resulting linear estimation problems. Thus, the linearisation potentially yields approximation errors that the filter does not consider in the prediction and update steps. The Extended Kalman filter's error estimates tend to underestimate state uncertainties. In order to improve that, the Unscented Kalman filter comes to this stage, which picks so-called sigma point samples from the filtering distribution and propagates/updates them through the nonlinear state and measurement models. The resulting weighted set of sigma points now represents the updated filtering distribution, which is then approximated as a moment match gaussian distribution. The results of the Unscented Kalman filter state estimates will represent the state uncertainty better than the estimates obtained from the Extended Kalman filter with an increased computational cost.*
